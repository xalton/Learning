## Utilisation Numpy

### Application sur le théorème central limite: 
Le théorème central limite établit la convergence en loi de la somme d'une suite de variables aléatoires vers la loi normale. Intuitivement, ce résultat affirme que toute somme de variables aléatoires indépendantes tend dans certains cas vers une variable aléatoire gaussienne.

```python
domaine = range(-100,100)
mu = 0
sigma = 20
f = lambda x: 1/(sqrt(2*pi*pow(sigma,2)))*exp(-pow((x-mu),2)/(2*pow(sigma,2)))
y = [f(x) for x in domaine]
plt.plot(domaine,y)
```

```python
matrice2 = np.random.rand(1000,10000)
somme2 = np.sum(matrice2,0)
print("Moyenne empirique de la distribution: {}".format(np.mean(somme2)))
print("Variance empirique de la distribution: {}".format(np.var(somme2)))


mu = 500
sigma = sqrt(84)
gauss = [8000*f(x) for x in range(460,540)]
plt.hist(somme2,bins=100,label = 'Distribution empirique')
plt.plot(range(460,540),gauss, label = 'Gausienne')
plt.legend()
plt.show()
```
