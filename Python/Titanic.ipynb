# Les nauffragés du Titanic
------------------------------------------

```python
#Modules
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.age.unique() #age possible
```

```python
titanic.describe(include='all')
```

```python
titanic = titanic.dropna() #élimine les NAN
titanic.head()
```

## Comment sont répartis les survivants ?

```python
titanic.pivot_table('survived',index='sex',columns='class')  # nombre moyen de survivant en fonction du sexe et de la classe
```

```python
titanic.pivot_table('survived',index='sex',columns='class',aggfunc='sum') #nombre total de survivants en fonction de ...
```

```python
titanic.dropna(inplace=True)
age = pd.cut(titanic['age'], [0, 18, 80]) #On regroupe en 3 catégories
titanic.pivot_table('survived', ['sex', age], 'class')
```
